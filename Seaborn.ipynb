```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# link : https://seaborn.pydata.org/
```

```python
tips = sns.load_dataset('tips')
tips
```

Data visualisation using seaborn on data set named tips

```python
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```python
#count plot
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
#box plot
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")

# it trepresent categorial data , here box represnt most of data and line between same colured boxes represnt mean
# above point outliers (excepional data (not take these line seriously ))
```

```python
 #box plot
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")
```

```python
#we obseve on every day males are spending more money and on sonday most feamales come 
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
 #on both side they are covering same area so using split you can merge (unwanted description soory , please see documentation )
 sns.violinplot(x='day',y='total_bill',data=tips,hue='sex', split=True)
```

###Distribution plot




```python
# Distribution plot allows you to show distribution of one variable 
```

```python
#the curvy line that you see here is KDE (kernel density estimation) , you can remove by making it false 
#bins is not just it is long blocks default it is 10


sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30,)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
# Join Plot : distribution between two plot , meaning you can combine two distribution plot in one graph  
```

```python
#default kind of join plot is scatter 
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='hex',color="red")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='kde',)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='reg',)
```

```python
# ? what if you want to caompare all the features and thier relation between all the features , for this we use pair plot
```

```python
#pair plot : it represent pair wise relation between all the features 
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```
